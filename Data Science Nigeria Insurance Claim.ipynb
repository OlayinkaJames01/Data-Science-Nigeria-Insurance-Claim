### Data Preprocessing

```python
def missing_values_tables(df):
    # Total Missing Values
    missing_values = df.isnull().sum()
    
    # Percentage of Missing Values
    missing_value_percent = 100 * (df.isnull().sum()/len(df))
    
    # Make a table with the results
    missing_value_table = pd.concat([missing_values, missing_value_percent], axis = 1)
    
    # Rename the columns 
    missing_values_table_rename_columns = missing_value_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values'})
                        
    # Sort the table by percentage of missing descending
    missing_values_table_rename_columns = missing_values_table_rename_columns[missing_values_table_rename_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
    
    # Printing some sumarry information
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(missing_values_table_rename_columns.shape[0]) +
              " columns that have missing values.")
                                   
     # Return the dataframe with missing information
    return missing_values_table_rename_columns
```

```python
data=pd.concat([train.drop('total_claim_amount',axis=1),test],axis=0)
```

```python
data.info()
```

```python
map_feat = ['insured_sex', 'insured_education_level', 'insured_occupation', 'insured_hobbies', 'insured_relationship', 
            'policy_state', 'property_damage', 'incident_type', 'collision_type', 'incident_severity', 'authorities_contacted', 
           'incident_state', 'incident_city', 'police_report_available', 'auto_make', 'auto_model', 'policy_deductable', 'number_of_vehicles_involved',
           'bodily_injuries', 'witnesses']

for col in map_feat:
    data[col] = data[col]
    
for col in map_feat:
    dummies = pd.get_dummies(data[col]).add_prefix(col+'_')
    data = pd.concat([data,dummies],axis=1).drop(col,axis=1)
```

### Stacking Models 

```python
predict1=model1.predict(X).reshape(-1,1)
predict2=model2.predict(X).reshape(-1,1)
predict3=model3.predict(X).reshape(-1,1)
predict4=model4.predict(X).reshape(-1,1)

pred1=model1.predict(test).reshape(-1,1)
pred2=model2.predict(test).reshape(-1,1)
pred3=model3.predict(test).reshape(-1,1)
pred4=model4.predict(test).reshape(-1,1)

X_new_train=np.concatenate([predict1, predict2, predict3,predict4],axis=1)

X_new_test=np.concatenate([pred1,pred2,pred3,pred4],axis=1)

modelf=XGBRegressor(learning_rate = 0.3, n_jobs = -1, n_estimators = 10, max_depth = 1).fit(X_new_train,y)
print('Stacked score:{}'.format(modelf.score(X_new_train,y)))
```

```python
modelf.fit(X_new_train,y)
```

```python
predictionf = modelf.predict(X_new_test)
```

```python
result = {'Customer_ID': test_df['Customer_ID'], 'total_claim_amount': predictionf}
results = pd.DataFrame(result)
```

```python
results.to_csv('stack1.csv', index = False)
```
